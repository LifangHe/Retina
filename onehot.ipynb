```python
import numpy as np
import h5py
import pickle
```

```python
name_onehot = pickle.load(open("name_onehot.p","rb"))

with h5py.File('num_arr.h5', 'a') as na:
    for arnum in range(10,83):
        print arnum, '...'
        imgnum = np.array(na.get('imgnum'+str(arnum)))
        rows = imgnum.shape[0]
        zeroarray = np.zeros((rows,4))
        for i in range(rows):
            value = name_onehot[str(imgnum[i]+'_T.txt')]
            zeroarray[i][value] = 1
        na.create_dataset("onehot"+str(arnum),data=zeroarray,compression="gzip",compression_opts=9)
        print 'saved', "onehot", str(arnum)
```
